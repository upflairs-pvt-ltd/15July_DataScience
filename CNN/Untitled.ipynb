```python
cat_dir_path = r"C:\Users\Ranjit\Desktop\15July_DataScience\CNN\Data\images\cats"
dog_dir_path = r"C:\Users\Ranjit\Desktop\15July_DataScience\CNN\Data\images\dogs"
```

```python
cats_images_ls = os.listdir(cat_dir_path)
dogs_images_ls = os.listdir(dog_dir_path)
```

```python
Image.open(os.path.join(cat_dir_path,cats_images_ls[5]))
```

```python
path_dict = {'cats':[],'dogs':[]} 

for filename in cats_images_ls:
    path_dict['cats'].append(os.path.join(cat_dir_path,filename))

for filename in dogs_images_ls:
    path_dict['dogs'].append(os.path.join(dog_dir_path,filename))
```

```python
# cv2.imread(r"C:\Users\Ranjit\Desktop\15July_DataScience\CNN\Data\images\cats\cat_1.jpg") 
```

```python
img_arr_ls = [] 
labels = [] 
for class_name , filepaths in path_dict.items(): 
    for filepath in filepaths: 
        img_arr = cv2.imread(filepath) 
        resized_img = cv2.resize(img_arr,(224,224))
        img_arr_ls.append(resized_img)
        labels.append(class_name)
```
